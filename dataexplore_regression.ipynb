```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
from subprocess import check_output

print(check_output(["ls","../datasets"]).decode("utf8"))
```

```python
data = pd.read_csv("../datasets/fifa-18-demo-player-dataset/CompleteDataset.csv")
data.head()
```

```python
# helper function to see column as .head() and .tail() for row. 
# to see the column val if the column number is way bigger than the cell size
def col_viewer(self, m, n):
    return self.iloc[:, m:n]

pd.DataFrame.col_viewer = col_viewer
```

```python
dat1 = pd.DataFrame(data)
dat1.col_viewer(0,20).head()
```

```python
dat1.col_viewer(20,40).head()
```

```python
# omit/drop the column that aren't necessary 
clean_dat = dat1.drop(['Photo', 'Flag', 'Club Logo'], axis=1)
clean_dat.head()
```

```python
df1 = pd.DataFrame(data)
df1 = df1[["Name","Age", "Nationality", "Overall", "Potential", "Club"]]
df1.head()
```

```python
df2 = pd.DataFrame(data)
df2 = df2.loc[:,"RB":"ST"]
df2.head()
```

```python
# concat two dataframe to cleanup data into what we need only
# this is only for exploration purpose :p
cldat = pd.concat([df1, df2], axis=1, sort=False)
cldat.head()
```

```python
# lets filter out the Goalkeeper database
# as you can see, all the goalkeeper value in 'RB' column is null
gk = clean_dat["RB"].isnull() == True
gk_db = clean_dat[gk]
gk_db.head()
```

```python
# now lets filter out the field player database
# it's simply a negation of GK function
plyr = clean_dat["RB"].isnull() != True
plyr_db = clean_dat[plyr]
plyr_db.head()
```
