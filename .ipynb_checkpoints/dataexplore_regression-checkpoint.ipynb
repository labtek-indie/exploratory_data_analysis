```python
data = pd.read_csv("../datasets/fifa-18-demo-player-dataset/CompleteDataset.csv")
data
```

```python
df1 = pd.DataFrame(data)
df1 = df1[["Name","Age", "Nationality", "Overall", "Potential", "Club"]]
df1.head()
```

```python
df2 = pd.DataFrame(data)
df2 = df2.loc[:,"RB":"ST"]
df2.head()
```

```python
# concat two dataframe to cleanup data into what we need only
cldat = pd.concat([df1, df2], axis=1, sort=False)
cldat.head()
```

```python
gk = cldat["RB"].isnull() == True
gk_db = cldat[gk]
gk_db.head()
```

```python
plyr = cldat["RB"].isnull() != True
plyr_db = cldat[plyr]
plyr_db.head()
```

```python
# helper function to see front and back of column as .head() and .tail() for row
def front(self, n):
    return self.iloc[:, :n]

def back(self, n):
    return self.iloc[:, -n:]

pd.DataFrame.front = front
pd.DataFrame.back = back

col_viewer = pd.DataFrame(sup_dat3)
```
